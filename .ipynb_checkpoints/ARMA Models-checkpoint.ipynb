```python
from pymongo import MongoClient
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import time

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

import warnings
warnings.filterwarnings('ignore')
```

```python
# connecting to the db
client=MongoClient(host="localhost", port=27017)
db=client["airquality"]
collection=db["nairobi"]
```

```python
def wrangle(collection, resample_rule='1H'):
    
    result=collection.find({"value_type":"P2"},
                          projection={"value":1, "timestamp":1,  "_id":0})
    
    df=pd.DataFrame(result).set_index("timestamp")
        
    df.index=pd.to_datetime(df.index)
    
    #renaming column
    df.rename(columns={"value":"P2"}, inplace=True)
    
  
    #localize time to nairobi
    df.index=df.index.tz_localize("UTC").tz_convert("Africa/Nairobi")  
        
    #according to Department of Health, the hazardous airquality is between 301-500
    df=df[df["P2"]<500] 
    
    #resampling data for hourly prediction
    y=df["P2"].resample("1H").mean()
    
    return y
```

```python
y=wrangle(collection, resample_rule="H1")
y.head()
```

```python
assert isinstance(y, pd.Series)
assert (y.isnull().sum()==0)
assert len(y) ==1505, f"'y' should have a  length of 1505 not {len(y)}"
```

# Explore

```python
# pacf and acf plots helps in determining the predictive power
fig, ax=plt.subplots(figsize=(15, 6))
plot_acf(y, ax=ax)
plt.xlabel("Lag [hours]")
plt.ylabel("Correlation Coefficient");
```

```python
fig, ax=plt.subplots(figsize=(15, 6))
plot_pacf(y, ax=ax)
plt.xlabel("Lag [hours]")
plt.ylabel("Correlation Coefficient");
```

# Split

```python
y_train=y.loc["2018-11-01":"2018-12-21"]
y_test=y.loc["2018-12-22":]
```

# Build Model

## Baseline

```python
y_train_mean=y_train.mean()
y_baseline=[y_train_mean] *len(y_train)
y_baseline_mae=mean_absolute_error(y_train, y_baseline)
print("Mean P2 Reading", round(y_train_mean, 2))
print("Baseline MAE", round(y_baseline_mae, 2))
```

ARMA models simply adds a moving average to the AR model wich considers the error made in the previous predictions.

AR: 
<font size="+1">$$y_{T} = \beta_{0} + \beta_{1} y_{T-1} $$


ARMA:
    
<font size="+1">$$y_{T} = \beta_{0} + \beta_{1} y_{T-1}  + \theta \epsilon_{T-1} $$

```python
# hyperparameter tuning.
p_params = range(0,25,8)
q_params = range(0,3,1)
```

```python
#to store scores
mae_score={}

for p in p_params:
    
    #create new key in the dict which will assign mae and time elapsed.
    mae_score[p]=list()
    
    for q in q_params:
        
        #order for ARMA not ARIMA
        order=(p, 0, q) 
        
        #begin timing the model.
        start_time=time.time()
        
        #train the model
        model=ARIMA(y_train, order=order).fit()  
        elapsed_time=round(time.time() - start_time, 2)
        print(f"ARIMA model in {order} for {elapsed_time} seconds")
        
        #generate in-sample predictions
        y_pred=model.predict()
        
        #calculating the training mae
        mae=mean_absolute_error(y_train, y_pred)
        
        
        mae_score[p].append((mae,elapsed_time))
```

```python
mae_df=pd.DataFrame(mae_score)
mae_df
```

```python
heatmap_val={}
for i in range(len(mae_score)):
    heatmap_val[i]=list()
    ele=list(mae_score.values())[i]    
    for t in range(len(ele)):
        heatmap_val[i].append(round(ele[t][0], 4))
```

```python
df=pd.DataFrame(heatmap_val)
df.rename(columns={0:"0",1:"8",2:"16",3:"24"}, inplace=True)
```

```python
sns.heatmap(df,cmap="pink")
plt.xlabel("p values")
plt.ylabel("q values")
plt.title('ARMA Grid Search (Criterion:MAE)');
```

The mae scores seems to be better as p values increases, particularly when p equals to 24 but it takes longer to train. I would like to have a balance of both, better performance and low training time.

```python
heatmap_val={}
for i in range(len(mae_score)):
    heatmap_val[i]=list()
    ele=list(mae_score.values())[i]    
    for t in range(len(ele)):
        heatmap_val[i].append(ele[t][1])
```

```python
df=pd.DataFrame(heatmap_val)
df.rename(columns={0:"0",1:"8",2:"16",3:"24"}, inplace=True)
```

```python
sns.heatmap(df,cmap="Blues")
plt.xlabel("p values")
plt.ylabel("q values")
plt.title('ARMA Grid Search (Elapsed Time)');
```

# Evaluate model

### walk-forward-validation

```python
y_predict_wfv=pd.Series([])
history=y_train.copy()
for i in range(len(y_test)):
    #model
    model=ARIMA(history, order=(16,0,2)).fit()
    next_pred=model.forecast()
    y_predict_wfv._append(next_pred)
    history=history._append(y_test[y_predict_wfv.index])
```

```python
test_mae=mean_absolute_error(y_test, y_predict_wfv)
print("Test MAE (walk forward validation):", round(test_mae, 2))
```

### Predictions and Real values

```python
df_predictions = pd.DataFrame({"y_test":y_test, "y_pred_wfv":y_pred_wfv})
fig = px.line(df_predictions, labels={"values":"PM2.5"})
fig.show()
```
